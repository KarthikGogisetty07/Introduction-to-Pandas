```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading a tmdb data from kaggle
df = pd.read_csv('tmdb_movies_data.csv')
```

```python
df.head()
```

```python
#considering few parameters that are requried!

my_subset = df[['id','popularity', 'original_title', 'genres', 'vote_count', 'vote_average','release_year']]
my_subset.head()
```

```python
my_subset.plot(kind='scatter',x='popularity',y='vote_average',color='yellowgreen')
plt.show()
```

```python
my_subset.plot(kind='scatter',x='vote_count',y='vote_average',color='purple')
plt.show()
```

```python
df1 = my_subset[my_subset['vote_average']>=6]
df1.head()
```

```python
df2 = my_subset[my_subset['popularity']>=2]
df2.head()
```

```python
### From the graphs the limits for a good movie suggestion based on popularity and vote_average(rating) are selected 
```

```python
Data = pd.merge(df1, df2)
Data.head()
```

```python
Data.sort_values('vote_average', ascending = False).head(10)
```

```python
Genre=input("Enter Your genre >> ")
Genre=Genre.capitalize()
```

```python
Suggested = Data[Data['genres'].str.contains(Genre)]
Suggested.sort_values('vote_average', ascending = False).head()
```
